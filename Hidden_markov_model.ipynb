```python
import numpy as np
```

```python
#Define matrices
A = np.array([[.5,.2,.3],[.2,.4,.4],[.4,.1,.5]])
B = np.array([[.8,.2],[.1,.9],[.5,.5]])
pi = np.array([[.5],[.3],[.2]])
observations = np.array([0,1,0,1])
```

```python
sequences = {}
sequence = ""
total_post = 0
for i in range(pi.shape[0]):
  for j in range(A.shape[0]):
    for k in range(A.shape[0]):
      for h in range(A.shape[0]):
          sequence = str(i) + str(j) + str(k) +str(h)
          probability = (pi[i][0]) * (A[i][j]) * (A[j][k]) * (A[k][h])
          likelyhood =  B[int(sequence[0])][0] * B[int(sequence[1])][1] * B[int(sequence[2])][0] * B[int(sequence[3])][1]
          total_post += probability*likelyhood
          #print(f"{sequence}: {probability}")
          sequences[sequence] = [probability,likelyhood, probability*likelyhood]

#sort sequences by prior probability
sequences = dict(sorted(sequences.items(), key=lambda item: item[1][2]))
for (K,V) in sequences.items():
  sequences[K].append((V[2])/total_post)

for item in sequences.items():
  print(item)

```
